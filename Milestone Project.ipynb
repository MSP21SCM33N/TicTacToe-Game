```python
board = [1,2,3,4,5,6,7,8,9]

def player_input(string_1, string_2):
    for numbers in board:
         if int(string_2) == numbers:
            element = board.index(numbers)
            board[element] = string_1
            print(board[element])
            continue
         else:
            continue

def print_func():
    print(board[0:3])
    print(board[3:6])
    print(board[6:9])

def game_checker(marker):
    if board[0] == board[1] == board[2] == marker:
        return 'Winner'
    elif board[0] == board[3] == board[6] == marker:
        return 'Winner'
    elif board[0] == board[4] == board[8] == marker:
        return 'Winner'
    elif board[3] == board[4] == board[5] == marker:
        return 'Winner'
    elif board[6] == board[7] == board[8] == marker:
        return 'Winner'
    elif board[2] == board[4] == board[6] == marker:
        return 'Winner'
    elif board[1] == board[4] == board[7] == marker:
        return 'Winner'
    elif board[2] == board[5] == board[8] == marker:
        return 'Winner'
    else:
        return 'Tie' 

def game_main():
    z = 1
    print('Welcome to the tic-tac-toe game!')
    while z <= len(board):
        z += 1
        string_1 = input('Enter X or 0: ')
        
        if string_1 != 'X' and string_1 != '0':
            print('Error')
            break
            
        print_func()
        string_2 = input('Enter the number on the board where you want to place your marker: ')
        player_input(string_1, string_2)
        marker_1 = 'X'
        marker_2 = '0'
        
        if game_checker(marker_1) == 'Winner':
            print(game_checker(marker_1))
            break
        elif game_checker(marker_2) == 'Winner':
            print(game_checker(marker_2))
            break
        else:
            if z == len(board):
                print(game_checker(marker_1))
                break
                

game_main()
```
