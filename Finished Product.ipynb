```python
from IPython.display import clear_output
board = ['1','2','3','4','5','6','7','8','9']
test_1 = 'X'
test_2 = 'Y'


def test():
    global board
    if game_checker(test_1) == 'Winner' or game_checker(test_2) == 'Winner':
        board = ['1','2','3','4','5','6','7','8','9']
    elif game_checker(test_1) == 'Tie' or game_checker(test_2) == 'Tie':
        board = ['1','2','3','4','5','6','7','8','9']
    else:
        pass
    return board

def display_board():
    clear_output()
    print('   |   |')
    print(' ' + board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[6] + ' | ' + board[7] + ' | ' + board[8])
    print('   |   |')


def player_input(string_1, string_2):
    for numbers in board:
         if string_2 == numbers:
            element = board.index(numbers)
            board[element] = string_1
            print(board[element])
            continue
         else:
            continue

def game_checker(marker):
    if board[0] == board[1] == board[2] == marker:
        return 'Winner'
    elif board[0] == board[3] == board[6] == marker:
        return 'Winner'
    elif board[0] == board[4] == board[8] == marker:
        return 'Winner'
    elif board[3] == board[4] == board[5] == marker:
        return 'Winner'
    elif board[6] == board[7] == board[8] == marker:
        return 'Winner'
    elif board[2] == board[4] == board[6] == marker:
        return 'Winner'
    elif board[1] == board[4] == board[7] == marker:
        return 'Winner'
    elif board[2] == board[5] == board[8] == marker:
        return 'Winner'
    else:
        return 'Tie' 

def game_main():
    z = 1
    while z <= len(board):
        z += 1
        display_board()
        string_1 = input('Enter X or 0: ').upper()
        
        if string_1 != 'X' and string_1 != '0':
            print('Error')
            break
            
        string_2 = input('Enter the number on the board where you want to place your marker: ')
        player_input(string_1, string_2)
        marker_1 = 'X'
        marker_2 = '0'
        
        if game_checker(marker_1) == 'Winner':
            print("Player 'X' is the Winner!")
            result = input('Do you want to play again? Y/N').upper()
            if result == 'Y':
                test()
                game_main()
            else:
                break
        elif game_checker(marker_2) == 'Winner':
            print("Player '0' is the Winner!")
            result = input('Do you want to play again? Y/N').upper()
            if result == 'Y':
                test()
                game_main()
            else:
                break
        else:
            if z == len(board):
                print('The game is a tie!')
                result = input('Do you want to play again? Y/N').upper()
                if result == 'Y':
                    test()
                    game_main()
                else:
                    break
        
game_main()
```
